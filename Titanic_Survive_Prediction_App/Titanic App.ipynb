```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import warnings 
from sklearn import *
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from tkinter import *
warnings.filterwarnings("ignore")
```

```python
def mod():
#   prediction function
    def survive_prob(new_input):
        x= pd.read_csv(r"C:\Users\david\Desktop\Titanic_Survive_Prediction_App\X_train.csv")
        y= pd.read_csv(r"C:\Users\david\Desktop\Titanic_Survive_Prediction_App\y_train.csv")
        x= x.drop(columns='Adult')
        model = LogisticRegression()
        model.fit(x,y)
        kfold = KFold(n_splits=10, random_state=None, shuffle=False)
        cross_val = cross_val_score(model, x, y, cv=kfold, scoring='accuracy')
        accuracy =cross_val.mean()
        predict_proba=model.predict_proba(new_input)[:,1]
        com=[predict_proba,accuracy]
        return com
    
    window = Tk()
    window.title("Predicting-Titanic-Survive-Probability")
    window.geometry('650x500')

    lbl = Label(window, text="Enter the following parameters")
    lbl.grid(column=2, row=0)
# form
    lbl = Label(window, text=" Age")
    lbl.grid(column=1, row=1)
    age = Entry(window,width=20)
    age.grid(column=3, row=1)
    age.focus()

    lbl = Label(window, text=" Gender (m=0,f=1)")
    lbl.grid(column=1, row=2)
    gender = Entry(window,width=20)
    gender.grid(column=3, row=2)
    gender.focus()

    lbl = Label(window, text=" Passenger Class (1-3)")
    lbl.grid(column=1, row=3)
    passengerClass = Entry(window,width=20)
    passengerClass.grid(column=3, row=3)
    passengerClass.focus()

    lbl = Label(window, text=" traveling companion")
    lbl.grid(column=1, row=4)
    travelingCompanion = Entry(window,width=20)
    travelingCompanion.grid(column=3, row=4)
    travelingCompanion.focus()

    lbl = Label(window, text="embarked on (Southampton=1, Cherboug=2, Quenstown=3)")
    lbl.grid(column=1, row=5)
    embarked = Entry(window,width=20)
    embarked.grid(column=3, row=5)
    embarked.focus()

    lbl = Label(window, text="Master title (yes=1, no=0)")
    lbl.grid(column=1, row=6)
    masterTitle = Entry(window,width=20)
    masterTitle.grid(column=3, row=6)
    masterTitle.focus()
# end of form

    def clicked():
#       Passenger Class setup
        firstClass = 0
        secondClass = 0
        thirdClass = 0
        if(int(passengerClass.get())==1):
            firstClass = 1
        elif(int(passengerClass.get())==2):
            secondClass = 1
        else:
            thirdClass = 1
#       Embarked setup
        Southamton = 0
        Cherbourg = 0
        Queenstown = 0
        if(int(embarked.get())==1):
            Southamton = 1
        elif(int(embarked.get())==2):
            Cherbourg = 1
        else:
            Queenstown = 1
        
        new_input=[int(gender.get()), int(age.get()),firstClass,secondClass,thirdClass,int(travelingCompanion.get()),Southamton, Cherbourg, Queenstown,int(masterTitle.get())]

        new_input= np.reshape(new_input,(1,-1))
        d=survive_prob(new_input)
        ans=d[0].item(0)
        sc=str(d[1].item(0)*100)
        
#       Rate probabilitas keselamatan
        if(ans>=0.5):
            s="High change of survive"
        else:
            s="Low change of survive"
        ot.configure(text="Result: " + s +"\n(prob: "+ str(ans) +"%)"+ "\n(accuracy ="+ sc +"% )" )


    btn = Button(window, text="submit",command=clicked)
    btn.grid(column=3, row=15)

    ot = Label(window, text="",font=("Arial Bold", 10))
    ot.grid(column=1, row=17)

    window.mainloop()
mod()    
```
